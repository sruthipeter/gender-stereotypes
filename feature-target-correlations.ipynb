```r
library(ggplot2)
## Function for arranging ggplots. use png(); arrange(p1, p2, ncol=1); dev.off() to save.
require(grid)
vp.layout <- function(x, y) viewport(layout.pos.row=x, layout.pos.col=y)
arrange_ggplot2 <- function(..., nrow=NULL, ncol=NULL, as.table=FALSE) {
	dots <- list(...)
	n <- length(dots)
	if(is.null(nrow) & is.null(ncol)) { nrow = floor(n/2) ; ncol = ceiling(n/nrow)}
	if(is.null(nrow)) { nrow = ceiling(n/ncol)}
	if(is.null(ncol)) { ncol = ceiling(n/nrow)}
        ## NOTE see n2mfrow in grDevices for possible alternative
grid.newpage()
pushViewport(viewport(layout=grid.layout(nrow,ncol) ) )
	ii.p <- 1
	for(ii.row in seq(1, nrow)){
	ii.table.row <- ii.row	
	if(as.table) {ii.table.row <- nrow - ii.table.row + 1}
		for(ii.col in seq(1, ncol)){
			ii.table <- ii.p
			if(ii.p > n) break
			print(dots[[ii.table]], vp=vp.layout(ii.table.row, ii.col))
			ii.p <- ii.p + 1
		}
	}
}
data <- read.csv(file='output.csv', head=TRUE)
options(repr.plot.width=8, repr.plot.height=5)
```

```r
# Age vs rest
spdm <- ggplot(data,aes(Age, y=spdm)) + geom_point() + ggtitle('Self Perceived Decision Making Power') + theme(plot.title = element_text(size=12, vjust=2), axis.title.y=element_text(size=10)) + ylab('Score')
ims <- ggplot(data,aes(Age, y=ims)) + geom_point() + ggtitle('Index of Marital Satisfaction') + theme(plot.title = element_text(size=12, vjust=2), axis.title.y=element_text(size=10)) + ylab('Score')
spi <- ggplot(data,aes(Age, y=SharedPowerIndex)) + geom_point() + ggtitle('Shared Power Index') + theme(plot.title = element_text(size=12, vjust=2), axis.title.y=element_text(size=10)) + ylab('Score')
dpi <- ggplot(data,aes(Age, y=DecisionPowerIndex)) + geom_point() + ggtitle('Decision Power Index') + theme(plot.title = element_text(size=12, vjust=2), axis.title.y=element_text(size=10)) + ylab('Score')
arrange_ggplot2(spdm,ims,spi,dpi,ncol=2)
```
